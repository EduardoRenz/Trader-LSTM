```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go


%matplotlib inline
```

```python
corretoras = pd.read_csv('CorBov.txt',header=None,names=['codigo','nome'],sep=';')

#historico = pd.read_csv("dolar/21022020/historico/DOLH20/_intraday05Min",sep=':',header=None)
negocios = pd.read_csv("dolar/21022020/20200221_dolh20",sep=';',header=None,names=['preco','qtd','datetime','comprador','vendedor','agressor'])
negocios = negocios.merge(corretoras,left_on="vendedor",right_on="codigo",how='left').rename(columns={"nome":"c_vendedor"})
negocios = negocios.merge(corretoras,left_on="comprador",right_on="codigo",how='left').rename(columns={"nome":"c_comprador"})
negocios = negocios.drop(['codigo_x','codigo_y','comprador','vendedor'],axis=1)
negocios.c_vendedor = negocios.c_vendedor.astype('category')
negocios.c_comprador = negocios.c_comprador.astype('category')
negocios.index =pd.to_datetime(negocios.datetime , format='%Y%m%d%H%M%S')# %%
```

```python

volume = negocios['qtd'].resample('10Min').sum()
```

```python
fig = make_subplots(rows=2,cols=1, shared_xaxes=True)
fig.add_trace(go.Scatter( x=negocios.index, y=negocios.preco),row=1,col=1)
fig.add_trace(go.Bar( x=volume.index, y=volume.values),row=2,col=1)
#fig = px.line(negocios, x=negocios.index, y="preco", title='Historico do dia',width=1400)
#fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
fig = px.bar(volume,x=volume.index,y='qtd')
fig.show()
```
